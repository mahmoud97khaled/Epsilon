# Merging tables

```python
import os, sys

# Open a file
path = "D:/Swimming App/DATA/Scrapped data/200 m"
dirs = os.listdir( path )
df = []
for n in range (len (dirs)) :
    Dir = 'D:/Swimming App/DATA/Scrapped data/200 m/'+dirs[n]
    df_storke = pd.read_csv(Dir)
    df_storke['Stroke'] = pd.Series([n for x in range(len(df_storke.index))])
    df.append(df_storke)
    result = pd.concat(df)
def Stroke (x):
    if x == 0:
        return 'Backstroke'
    elif x == 1:
        return 'Breaststroke'
    elif x == 2:
        return 'Butterfly'
    else :
        return 'Freestyle'
result['Stroke'] = result['Stroke'].apply(Stroke)
def gender (x):
    if x == 0:
        return 'm'
    else :
        return 'f'
result['gender'] = result['gender'].apply(gender)
df
result.drop('Unnamed: 0', axis = 1, inplace =True)
result.to_csv('D:/Swimming App/200m_final.csv', index=False)
result
```
